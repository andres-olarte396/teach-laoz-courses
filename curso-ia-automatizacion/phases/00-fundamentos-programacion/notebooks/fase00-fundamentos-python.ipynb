## 🔄 Sección 4: Estructuras de Control - Loops y Condicionales

Aprende a controlar el flujo de tu programa con condiciones y repeticiones.

```python
# 🔀 CONDICIONALES - Tomar decisiones en el código

# Sistema de calificaciones
def evaluar_nota(nota):
    """Evalúa una nota y devuelve la categoría correspondiente."""
    if nota >= 90:
        return "Excelente (A)"
    elif nota >= 80:
        return "Muy bueno (B)"
    elif nota >= 70:
        return "Bueno (C)"
    elif nota >= 60:
        return "Regular (D)"
    else:
        return "Necesita mejorar (F)"

# Probar diferentes notas
notas_estudiantes = [95, 87, 73, 65, 45, 92]

print("🎓 EVALUACIÓN DE NOTAS:")
for i, nota in enumerate(notas_estudiantes, 1):
    categoria = evaluar_nota(nota)
    print(f"Estudiante {i}: {nota} puntos - {categoria}")

# Condicional con múltiples criterios
edad = 20
tiene_licencia = True
tiene_auto = False

print(f"\n🚗 EVALUACIÓN PARA CONDUCIR:")
print(f"Edad: {edad}, Licencia: {tiene_licencia}, Auto: {tiene_auto}")

if edad >= 18 and tiene_licencia:
    if tiene_auto:
        mensaje = "✅ Puede conducir su propio auto"
    else:
        mensaje = "⚠️ Puede conducir, pero necesita un auto"
else:
    mensaje = "❌ No puede conducir"

print(f"Resultado: {mensaje}")

# Operador ternario (condicional en una línea)
estado = "Aprobado" if nota >= 70 else "Reprobado"
print(f"\nEstado final: {estado}")
```

```python
# 🔄 BUCLES - Repetir acciones eficientemente

print("🔢 BUCLE FOR - Repeticiones con rango conocido:")
# Tabla de multiplicar del 7
numero = 7
print(f"Tabla del {numero}:")
for i in range(1, 11):
    resultado = numero * i
    print(f"{numero} x {i} = {resultado}")

print(f"\n📋 BUCLE FOR - Iterar sobre listas:")
# Procesar lista de temperaturas
temperaturas = [22, 25, 19, 30, 27, 24, 21]
dias = ["Lunes", "Martes", "Miércoles", "Jueves", "Viernes", "Sábado", "Domingo"]

for dia, temp in zip(dias, temperaturas):
    if temp > 25:
        estado = "🔥 Caluroso"
    elif temp < 20:
        estado = "❄️ Frío"
    else:
        estado = "🌡️ Agradable"
    print(f"{dia}: {temp}°C - {estado}")

print(f"\n🎯 BUCLE WHILE - Repetir mientras se cumpla condición:")
# Simulador de ahorro
dinero_ahorrado = 0
meta = 1000
mes = 0
ahorro_mensual = 150

print(f"Meta de ahorro: ${meta}")
print(f"Ahorro mensual: ${ahorro_mensual}")
print("\\nProgreso:")

while dinero_ahorrado < meta:
    mes += 1
    dinero_ahorrado += ahorro_mensual
    porcentaje = (dinero_ahorrado / meta) * 100
    print(f"Mes {mes}: ${dinero_ahorrado} ({porcentaje:.1f}% de la meta)")

print(f"\\n🎉 ¡Meta alcanzada en {mes} meses!")
```

## 🔧 Sección 5: Funciones - Organizar y Reutilizar Código

Las funciones son bloques de código reutilizable que hacen nuestros programas más organizados y eficientes.

```python
# 🔧 FUNCIONES - Bloques de código reutilizable

def saludar(nombre, edad=None):
    """
    Función que saluda a una persona.
    
    Args:
        nombre (str): Nombre de la persona
        edad (int, optional): Edad de la persona
    
    Returns:
        str: Mensaje de saludo personalizado
    """
    if edad:
        return f"¡Hola {nombre}! Tienes {edad} años."
    else:
        return f"¡Hola {nombre}!"

def calcular_imc(peso, altura):
    """
    Calcula el Índice de Masa Corporal (IMC).
    
    Args:
        peso (float): Peso en kilogramos
        altura (float): Altura en metros
    
    Returns:
        tuple: IMC y categoría correspondiente
    """
    imc = peso / (altura ** 2)
    
    if imc < 18.5:
        categoria = "Bajo peso"
    elif imc < 25:
        categoria = "Peso normal"
    elif imc < 30:
        categoria = "Sobrepeso"
    else:
        categoria = "Obesidad"
    
    return imc, categoria

def analizar_numeros(*numeros):
    """
    Analiza una lista variable de números.
    
    Args:
        *numeros: Números a analizar
    
    Returns:
        dict: Estadísticas de los números
    """
    if not numeros:
        return {"error": "No se proporcionaron números"}
    
    return {
        "cantidad": len(numeros),
        "suma": sum(numeros),
        "promedio": sum(numeros) / len(numeros),
        "minimo": min(numeros),
        "maximo": max(numeros)
    }

# Usar las funciones
print("👋 FUNCIÓN DE SALUDO:")
print(saludar("Ana"))
print(saludar("Carlos", 28))

print(f"\n⚖️ CALCULADORA DE IMC:")
imc, categoria = calcular_imc(70, 1.75)
print(f"IMC: {imc:.2f} - Categoría: {categoria}")

print(f"\n📊 ANÁLISIS DE NÚMEROS:")
stats = analizar_numeros(10, 25, 30, 15, 40, 35, 20)
for clave, valor in stats.items():
    if isinstance(valor, float):
        print(f"{clave.title()}: {valor:.2f}")
    else:
        print(f"{clave.title()}: {valor}")
```

### 💪 Ejercicio Práctico 2: Sistema de Gestión de Estudiantes

Vamos a crear un sistema completo que combine todo lo que hemos aprendido.

```python
# 🎓 PROYECTO INTEGRADOR: Sistema de Gestión de Estudiantes

class GestorEstudiantes:
    """Clase para gestionar información de estudiantes."""
    
    def __init__(self):
        self.estudiantes = {}
        self.contador_id = 1
    
    def agregar_estudiante(self, nombre, edad, carrera):
        """Agrega un nuevo estudiante al sistema."""
        id_estudiante = f"EST{self.contador_id:03d}"
        self.estudiantes[id_estudiante] = {
            "nombre": nombre,
            "edad": edad,
            "carrera": carrera,
            "notas": [],
            "activo": True
        }
        self.contador_id += 1
        return id_estudiante
    
    def agregar_nota(self, id_estudiante, materia, nota):
        """Agrega una nota a un estudiante."""
        if id_estudiante in self.estudiantes:
            self.estudiantes[id_estudiante]["notas"].append({
                "materia": materia,
                "nota": nota
            })
            return True
        return False
    
    def calcular_promedio(self, id_estudiante):
        """Calcula el promedio de un estudiante."""
        if id_estudiante in self.estudiantes:
            notas = self.estudiantes[id_estudiante]["notas"]
            if notas:
                total = sum(n["nota"] for n in notas)
                return total / len(notas)
        return 0
    
    def generar_reporte(self):
        """Genera un reporte completo de todos los estudiantes."""
        print("=" * 60)
        print("🎓 REPORTE ACADÉMICO COMPLETO")
        print("=" * 60)
        
        for id_est, info in self.estudiantes.items():
            promedio = self.calcular_promedio(id_est)
            estado = "✅ Aprobado" if promedio >= 70 else "❌ Reprobado"
            
            print(f"\\n📋 ID: {id_est}")
            print(f"👤 Nombre: {info['nombre']}")
            print(f"🎂 Edad: {info['edad']} años")
            print(f"📚 Carrera: {info['carrera']}")
            print(f"📊 Promedio: {promedio:.2f}")
            print(f"🎯 Estado: {estado}")
            
            if info['notas']:
                print("📝 Notas por materia:")
                for nota_info in info['notas']:
                    print(f"   • {nota_info['materia']}: {nota_info['nota']}")
            
            print("-" * 40)

# Crear el sistema y agregar datos de prueba
gestor = GestorEstudiantes()

# Agregar estudiantes
id1 = gestor.agregar_estudiante("Ana García", 20, "Ingeniería en Sistemas")
id2 = gestor.agregar_estudiante("Carlos López", 22, "Administración")
id3 = gestor.agregar_estudiante("Elena Martín", 19, "Psicología")

# Agregar notas
gestor.agregar_nota(id1, "Python", 95)
gestor.agregar_nota(id1, "Matemáticas", 87)
gestor.agregar_nota(id1, "Base de Datos", 92)

gestor.agregar_nota(id2, "Contabilidad", 78)
gestor.agregar_nota(id2, "Marketing", 85)
gestor.agregar_nota(id2, "Finanzas", 82)

gestor.agregar_nota(id3, "Psicología General", 88)
gestor.agregar_nota(id3, "Estadística", 76)
gestor.agregar_nota(id3, "Neurociencia", 91)

# Generar reporte
gestor.generar_reporte()
```

## 🎯 Proyecto Final: Analizador de Datos Personales

Vamos a crear un proyecto que combine todos los conceptos aprendidos en esta fase.

```python
# 🎯 PROYECTO FINAL: Analizador de Hábitos Personales

class AnalizadorHabitos:
    """Clase para analizar y hacer seguimiento de hábitos personales."""
    
    def __init__(self, nombre_usuario):
        self.nombre_usuario = nombre_usuario
        self.habitos = {}
        self.fecha_inicio = datetime.now()
    
    def agregar_habito(self, nombre_habito, objetivo_semanal):
        """Agrega un nuevo hábito a seguir."""
        self.habitos[nombre_habito] = {
            "objetivo_semanal": objetivo_semanal,
            "registros": [],
            "total_completado": 0
        }
        print(f"✅ Hábito '{nombre_habito}' agregado con objetivo de {objetivo_semanal} veces por semana")
    
    def registrar_actividad(self, nombre_habito, fecha=None):
        """Registra que se completó un hábito en una fecha específica."""
        if fecha is None:
            fecha = datetime.now().date()
        
        if nombre_habito in self.habitos:
            self.habitos[nombre_habito]["registros"].append(fecha)
            self.habitos[nombre_habito]["total_completado"] += 1
            print(f"🎯 Registrado: {nombre_habito} el {fecha}")
        else:
            print(f"❌ Hábito '{nombre_habito}' no encontrado")
    
    def calcular_estadisticas(self, nombre_habito):
        """Calcula estadísticas de un hábito específico."""
        if nombre_habito not in self.habitos:
            return None
        
        habito = self.habitos[nombre_habito]
        total_dias = (datetime.now().date() - self.fecha_inicio.date()).days + 1
        total_semanas = total_dias / 7
        
        promedio_semanal = habito["total_completado"] / total_semanas if total_semanas > 0 else 0
        porcentaje_objetivo = (promedio_semanal / habito["objetivo_semanal"]) * 100 if habito["objetivo_semanal"] > 0 else 0
        
        return {
            "total_completado": habito["total_completado"],
            "objetivo_semanal": habito["objetivo_semanal"],
            "promedio_semanal": promedio_semanal,
            "porcentaje_objetivo": porcentaje_objetivo,
            "total_dias_seguimiento": total_dias
        }
    
    def generar_reporte_completo(self):
        """Genera un reporte completo de todos los hábitos."""
        print("=" * 60)
        print(f"📊 REPORTE DE HÁBITOS - {self.nombre_usuario.upper()}")
        print("=" * 60)
        print(f"📅 Período de seguimiento: {(datetime.now().date() - self.fecha_inicio.date()).days + 1} días")
        
        if not self.habitos:
            print("\\n❌ No hay hábitos registrados")
            return
        
        for nombre_habito, datos in self.habitos.items():
            stats = self.calcular_estadisticas(nombre_habito)
            
            print(f"\\n🎯 HÁBITO: {nombre_habito.upper()}")
            print(f"   Objetivo semanal: {stats['objetivo_semanal']} veces")
            print(f"   Total completado: {stats['total_completado']} veces")
            print(f"   Promedio semanal: {stats['promedio_semanal']:.1f} veces")
            print(f"   Progreso del objetivo: {stats['porcentaje_objetivo']:.1f}%")
            
            # Determinar estado del hábito
            if stats['porcentaje_objetivo'] >= 100:
                estado = "🏆 ¡Excelente! Superando objetivos"
            elif stats['porcentaje_objetivo'] >= 80:
                estado = "✅ Muy bien, casi en objetivo"
            elif stats['porcentaje_objetivo'] >= 60:
                estado = "⚠️ Regular, necesita más constancia"
            else:
                estado = "❌ Crítico, requiere más esfuerzo"
            
            print(f"   Estado: {estado}")
            print("-" * 40)

# Crear analizador y simular datos
analizador = AnalizadorHabitos("Ana García")

# Agregar hábitos
analizador.agregar_habito("Ejercicio", 5)  # 5 veces por semana
analizador.agregar_habito("Lectura", 7)    # Diario
analizador.agregar_habito("Meditación", 3) # 3 veces por semana

# Simular registros de actividades (últimas 2 semanas)
from datetime import date, timedelta

# Ejercicio: 8 veces en 2 semanas
fechas_ejercicio = [
    date.today() - timedelta(days=13),
    date.today() - timedelta(days=11),
    date.today() - timedelta(days=9),
    date.today() - timedelta(days=7),
    date.today() - timedelta(days=5),
    date.today() - timedelta(days=3),
    date.today() - timedelta(days=2),
    date.today() - timedelta(days=1)
]

for fecha in fechas_ejercicio:
    analizador.registrar_actividad("Ejercicio", fecha)

# Lectura: 10 veces en 2 semanas
for i in range(10):
    fecha = date.today() - timedelta(days=random.randint(1, 14))
    analizador.registrar_actividad("Lectura", fecha)

# Meditación: 4 veces en 2 semanas
for i in range(4):
    fecha = date.today() - timedelta(days=random.randint(1, 14))
    analizador.registrar_actividad("Meditación", fecha)

# Generar reporte final
analizador.generar_reporte_completo()
```
